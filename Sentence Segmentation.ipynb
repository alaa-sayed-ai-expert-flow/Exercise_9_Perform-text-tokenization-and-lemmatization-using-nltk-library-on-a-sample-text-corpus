```python
from spacy.language import Language
```

```python
@Language.component("set_custom_boundaries")
def set_custom_boundaries(doc):
    for token in doc[:-1]:
        if token.text == ";":
            doc[token.i + 1].is_sent_start = True
    return doc

nlp.add_pipe("set_custom_boundaries", before="parser")
nlp.pipe_names
```

```python
string = '"Management is doing things right; leadership is doing the right things." -Alaa Sayed'
```

```python
doc_2 = nlp(string)
for sent in doc_2.sents:
    print(sent)
```

# Using NLTK

```python
string = '"Management is doing things right; leadership is doing the right things." -Alaa Sayed'
```

```python
from nltk.tokenize import sent_tokenize
```

```python
for s in sent_tokenize(string):
    print(s)
    print("--------------------------------------")
```

```python
from nltk.tokenize import PunktSentenceTokenizer
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(string)
tokenized = custom_sent_tokenizer.tokenize(string)
```

```python
print(tokenized)
```
