# First Simple app in NLP

```python
import spacy
```

```python
#    
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("Tesla is looking at buying U.S startup for $6 million dollars")
```

```python
for token in doc:
    print(token.text)
    print(token.shape)
    print(token.shape_)
    print(token.is_alpha) # is alphabetic or not --       
    print(token.is_stop) # is stop word or not
    print("========================================")
```

```python
string = "We're moving to L.A!"
```

```python
doc_2 = nlp(string)
for token in doc_2:
    print(token , end = " | ")
```
